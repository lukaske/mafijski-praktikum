## Uvod

Enodimenzionalni linearni harmonski oscilator (delec mase $m$
s kinetično energijo $T(p)=p^2/2m$ v kvadratičnem potencialu
$V(q)=m\omega^2 q^2/2$) opišemo z brezdimenzijsko Hamiltonovo funkcijo
\begin{equation*}
  H_0 = {1\over 2} \left( p^2 + q^2 \right) \>,
\end{equation*}
tako da energijo merimo v enotah $\hbar\omega$, gibalne količine
v enotah $(\hbar m\omega)^{1/2}$ in dolžine v enotah $(\hbar/m\omega)^{1/2}$.
Lastna stanja $|n\rangle$ nemotenega Hamiltonovega operatorja $H_0$ so
\begin{equation*}
  |n\rangle = (2^n n! \sqrt{\pi})^{-1/2} \mathrm{e}^{-q^2/2}\,  {\cal H}_n (q)\>,
\end{equation*}
kjer so ${\cal H}_n$ Hermitovi polinomi.
Lastne funkcije zadoščajo stacionarni Schrodingerjevi enačbi
\begin{equation*}
H_0 | n^0 \rangle = E_n^0 | n^0 \rangle
\end{equation*}
z nedegeneriranimi lastnimi energijami $E_n^0 = n + 1/2$
za $n=0,1,2,\ldots~$.  Matrika $\langle i | H_0 | j\rangle$
z $i,j=0,1,2,\ldots,N-1$ je diagonalna, z vrednostmi
$\delta_{ij}(i + 1/2)$ po diagonali.  Nemoteni Hamiltonki
dodamo anharmonski člen
\begin{equation*}
H = H_0 + \lambda q^4 \>.
\end{equation*}
Matriko s tem potencialom v nadaljevanju navajamo s $H$. Tedaj matriko zapišemo kot
$$
\begin{equation*}
  \langle  i | H | j \rangle = \langle  i | H_0 | j \rangle + \lambda \langle  i | q^4 | j \rangle
\end{equation*}
$$

Za izračun drugega člena si lahko pomagamo z naslednjimi zvezami
$$
\begin{align*}
  &\langle i | q | j \rangle = {1\over 2} \sqrt{i+j+1}\,\, \delta_{|i-j|,1} \\
  &\langle i|q^2|j\rangle
  = {1\over 2} \biggl[
    {\sqrt{j(j-1)}} \, \delta_{i,j-2}
  + {(2j+1)} \, \delta_{i,j}
  + {\sqrt{(j+1)(j+2)}} \, \delta_{i,j+2} \biggr] \\
  &\langle i|q^4|j\rangle
  = {1\over 2^4}\sqrt{2^i \, i!\over 2^{j} \, j! } \, \biggl[ \,
  \delta_{i,j+4} + 4\left(2j+3\right) \delta_{i,j+2}
                      + 12 \left(2j^2+2j+1\right) \, \delta_{i,j} \\[3pt]
  &+ 16j \left(2j^2-3j+1\right) \, \delta_{i,j-2}
     + 16j\left(j^3-6j^2+11j-6\right) \, \delta_{i,j-4} \biggr]
\end{align*}
$$
Sedaj rešujemo matrični sistem
\begin{equation*}
  H | n \rangle = E_n | n \rangle \>,
\end{equation*}
katerega bomo reševali v bazi lastnih funkcij harmonskega oscilatorja $| i \rangle$:
\begin{equation*}
  | n \rangle = \sum_{i=0}^N c_i | i \rangle
\end{equation*}

Metodo lahko nato prenesemo tudi na drugačne potenciale. V dodatku je analiziran še kvantni harmonski oscilator z dvema minimumoma, katerega $H$ zapišemo kot
\begin{equation*}
  H = {p^2\over 2} - 2q^2 + {q^4\over 10} \>.
\end{equation*}
Podobno kot prej lahko matriko zapišemo kot
\begin{equation*}
  \langle  i | H | j \rangle = \langle  i | H_0 | j \rangle - \frac{5}{2} \langle  i | q^2 | j \rangle + \frac{1}{10} \langle  i | q^4 | j \rangle
\end{equation*}
Matriko s tem potencialom v nadaljevanju navajamo s $H_2$.

## Hamiltonov matrični operator za an-harmonski oscilator

Za implementacijo sledimo formulam navedenim v uvodu. Perturbacijski prispevek anharmonskega potenciala si lahko, kot že prej navedeno, interpretiramo na 4 različne načine. Načine primerjamo po hitrosti izračuna. 

```python
# Function definitions
import timeit

def H_0(N, h0=True):
    H_0_m = np.zeros((N,N))
    if not h0:
        return H_0_m
    np.fill_diagonal(H_0_m, np.arange(0,N) + 0.5)
    return H_0_m

def get_ij(N):
    x = np.arange(0, N)
    y = np.arange(0, N)
    iv, jv = np.meshgrid(x, y, indexing='ij')
    i = iv.flatten().astype(np.float64)
    j = jv.flatten().astype(np.float64)
    return i, j

def q(N):
    i, j = get_ij(N)
    sum = (np.abs(i-j) == 1) * np.sqrt(j+i+1) * 0.5
    sum = sum.reshape((N,N))
    return sum

def q_2(N):
    i, j = get_ij(N)
    sum = ( (j-2==i) * np.sqrt(j*(j-1)) + (j==i) * (2*j+1) + (j+2==i) * np.sqrt((j+1)*(j+2)) ) * 0.5
    sum = sum.reshape((N,N))
    return sum

def fast_factorial(N):
    fact = np.zeros((N, N))
    for j_ in range(N):
        for i_ in range(N):
            if i_ == j_:
                fact[i_, j_] = 1.0
            elif i_ < j_:
                fact[i_, j_] = 1 / np.prod(np.arange(i_+1, j_+1), dtype=np.float64)
            else:
                fact[i_, j_] = np.prod(np.arange(j_+1, i_+1), dtype=np.float64)
    return fact.flatten()

def q_4(N):
    i, j = get_ij(N)
    sum = (j+4==i) + 4*(2*j+3) * (j+2==i) + 12*(2*j**2+2*j+1) * (j==i) + 16*j*(2*j**2-3*j+1) * (j-2==i) + 16*j*(j**3-6*j**2+11*j-6) * (j-4==i)
    sum *= np.sqrt(2**(i-j-8) * fast_factorial(N))
    sum = sum.reshape((N,N))
    return sum

def get_H(N, lam, type='q', h0 = True):
    if type == 'q':
        return H_0(N, h0) + lam * np.linalg.matrix_power(q(N), 4)
    elif type == 'q2':
        return H_0(N, h0) + lam * np.linalg.matrix_power(q_2(N), 2)
    elif type == 'q4':
        return H_0(N, h0) + lam * q_4(N)
    elif type == 'double':
        return H_0(N, h0) + 0.1 * np.linalg.matrix_power(q(N), 4) - 2.5 * np.linalg.matrix_power(q(N), 2)
    else:
        raise ValueError('Invalid type')

    
```

```python
t = {}
for type in ['q', 'q2', 'q4']:
    for n in range(0, 101):
        f = lambda: get_H(n, 1, type, False)
        reps = 5
        execution_time = timeit.timeit(f, number=reps)
        t_q = t.get(type, [])
        t_q.append(execution_time / reps)
        t[type] = t_q

plt.title("Povprečni čas izračuna matrike $H$ za različne velikosti matrike")
plt.plot(t['q'], label='$ \langle i | q | j \\rangle^4 $')
plt.plot(t['q2'], label='$ \langle i | q^2 | j \\rangle^2 $')
plt.plot(t['q4'], label='$ \langle i | q^4 | j \\rangle $')
plt.yscale('log')
plt.xlabel('N')
plt.ylabel('Čas (log) [s]')
plt.legend()
plt.show()
```

## Numerična diagonalizacija matrik

Za reševanje matrične enačbe sedaj potrebujemo učinkovit postopek za diagonalizacijo matrik. V primerjavo bomo vzeli diagonalizacijo po Jacobijevi metodi v lastni implementaciji (Python/numpy in C++), numpy metodo `numpy.linalg.eig` in metodo `Eigen::SelfAdjointEigenSolver` iz paketa Eigen za C++.

Metode bomo testirali s pomočjo generiranih naključnih matrik velikosti $2\times2$ do $100 \times 100$. Primerjali bomo hitrost in natančnost diagonalizacije (s pomočjo kvadrata ne-diagonalne Frobeniusove norme $P_M = \lVert M \rVert_F^2 \; - \sum_{i=1}^{n} M_{ii}^2$). 

---

#### Jacobijev algoritem

Za primerjavo z metodami v programskih paketih bomo "na roke" implementirali [Jacobijev algoritem za izračun lastnih vrednosti in lastnih vektorjev](https://en.wikipedia.org/wiki/Jacobi_eigenvalue_algorithm). To storimo tako v "enostavnem" programskem jeziku `Python` (z uporabo knjižnice `numpy`, saj je to najbolj realen praktičen primer) in v "hitrem" programskem jeziku `C++`. 

<quote>Gre za iterativno metodo, ki jo lahko skiciramo v naslednjih korakih:

1. Imamo simetrično matriko $S$ in Givensovo rotacijsko matriko $G = G(i,j,\theta)$. Tedaj je $S' = G^\top S \; G$ simetrična in podobna $S$.
2. (Zaradi optimizacije) si izberemo po absolutni vrednosti največji člen $S_{ij}$, ki ga imenujemo $\tt{pivot}$. Sestavimo tako rotacijsko matriko $G$, da je $S'_{ij} = S'_{ji} = 0$. To ponovimo $N = n(n-1)/2$-krat ($n$ je velikost matrike in $N$ število ne-diagonalnih členov). Če na matriko apliciramo $N$ Givensovih/Jacobijevih rotacij, takšno skupinsko operacijo definiramo kot $\tt{sweep}$.
3. Ponovimo poljubno mnogo operacij $\tt{sweep}$, da dosežemo željeno numerično natančnost ($P$). Natančnost merimo kot $ P = \lVert S' \rVert_F^2 \; - \sum_{i=1}^{n} S_{ii}^2$.

---

Implementacijo testiramo in ugotovimo, da ustrezno diagonalizira matriko.</quote>


```python
# Generate test samples

def generate_mtrix(n):
    eigenvalues = np.arange(1, n+1, 1)
    eigenvalues = np.random.permutation(eigenvalues) * np.random.rand()
    non_diagonal_matrix = np.diag(eigenvalues).astype(np.float64)
    for i in range(n):
        for j in range(i + 1, n):
            if (j!=i):
                val = np.random.rand() * n
                non_diagonal_matrix[i, j] = val
                non_diagonal_matrix[j, i] = val
    return non_diagonal_matrix

max_N = 100
matrices = []

for n in range(2, max_N):
    generated = generate_mtrix(n)
    matrices.append([n, generated])

df = pd.DataFrame(matrices, columns=['N', 'flattened_matrix'])
df.to_json('test_data.json', index=False)
```

```python
# Jacobi diagonalziation imlementation python
def off_diag_norm(A):
    A_ = A.copy()
    dia = np.diag_indices(A_.shape[0])
    dia_sum = np.sum(np.square(A_[dia])) # element-wise square
    return np.sum(np.square(A_)) - dia_sum # frobenius norm - diagonal sum

def find_pivot(A):
    n = A.shape[0]
    A_ = A.copy()
    diag = np.diag_indices(n)
    A_[diag] = 0
    max_idx = np.argmax(np.abs(A_.flatten()))
    p = max_idx // n
    q = max_idx % n
    return (p, q)

def Jacobi(A, precision=1e-10, max_iter=5):
    n = A.shape[0]
    A_ = A.copy()
    D_ = np.eye(n)
    S_val = off_diag_norm(A_)
    i = 0
    #print(f'Starting S = {S}')
    while (i < max_iter) and (S_val > precision):
        for p_index in range(n-1):
            for q_index in range(p_index+1, n):
                p, q = find_pivot(A_)
                if np.abs(A_[p,q]) < 1e-10:
                    t = 0
                else:
                    theta = (A_[q,q] - A_[p,p]) / (2*A_[p,q])
                    t = np.sign(theta) / (np.abs(theta) + np.sqrt(theta**2 + 1))
                c = 1 / np.sqrt(t**2 + 1)
                s = t * c
                J = np.eye(n)
                J[p,p] = c
                J[q,q] = c
                J[p,q] = s
                J[q,p] = -s
                D_ = D_ @ J
                A_ = J.T @ A_ @ J
        S_val = off_diag_norm(A_)
        #print(f'Iteracija {i}, S = {S}')
        i += 1
        #print(f'iter {i} reached for {n}, best S = {S_val}')
    #print('Best S = ', S_val, ' for ', n, ' iterations ', i)
    return A_, D_
```

Različne implementacije algoritmov za diagonalizacijo lahko sedaj primerjamo po hitrosti izvedbe za vsako peto matriko v naboru 100 matrik. Implementacijo v C++ smo v strojni jezik prevedli s komando in parametri `g++ -O3 -std=c++11 -I ./Eigen eigenDiag.cpp -o testEigen`.

```python
def benchmark_numpy(matrices):
    times = []
    N = []
    num_precision = []
    total_time = 0
    diagonalized = []
    for matrix in matrices:
        start = timeit.default_timer()
        _V, _M = np.linalg.eig(matrix)
        end = timeit.default_timer()
        t = (end - start) * 1000
        times.append(t)
        N.append(matrix.shape[0])
        num_precision.append(off_diag_norm(_M))
        diagonalized.append(_M)
    total_time = np.cumsum(times)
    return [N, times, num_precision, total_time]

def benchmark_jacobi(matrices):
    times = []
    N = []
    num_precision = []
    diagonalized = []
    for matrix in matrices:
        start = timeit.default_timer()
        A_, _D = Jacobi(matrix, precision=float(matrix.shape[0]), max_iter=10)
        end = timeit.default_timer()
        t = (end - start) * 1000
        times.append(t)
        N.append(matrix.shape[0])
        num_precision.append(off_diag_norm(A_))
        diagonalized.append(A_)
    total_time = np.cumsum(times)
    return [N, times, num_precision, total_time]
```

## **An**harmonski oscilator

Sedaj z diagonalizacijo rešimo perturbiran Hamiltonjan za anharmonski oscilator. Zavedamo se, da ima Hamiltonjan dejansko neskončno lastnih funkcij/vektorjev $ | n \rangle $ in lastnih vrednosti $E_n$. Torej lahko v naši simulaciji obravnavam samo približek z (dovolj) velikim $n$.

```python
def diagonalize(M):
    _v, _D = np.linalg.eig(M)
    sorting = np.argsort(_v)
    _v = _v[sorting]
    _D = _D[:, sorting]
    return _v, _D

n = np.arange(1, 500, 5)
H_0s = [H_0(n_) for n_ in n]
Hs = [diagonalize(get_H(n_ + 3, 1, 'q')[:n_, :n_]) for n_ in n] # 0 - eigenvalues, 1 - eigenvectors
```

```python
En = [eig[0][i*5] for i, eig in enumerate(Hs)]
En_later = [eig[0][i*5] for i, eig in enumerate(Hs[50:])]
En_late = [eig[0][i*5] for i, eig in enumerate(Hs[25:])]

fig, (ax2, ax1) = plt.subplots(nrows=2, ncols=1, sharex=True)

ax1.plot(n, En, label='Anharmonski oscilator, pad: 0')
ax1.plot(n[:75], En_late, label='Anharmonski oscilator, pad: 25', linestyle='dotted')
ax1.plot(n[:50], En_later, label='Anharmonski oscilator, pad: 50', linestyle='dotted')
ax1.plot(n, n + 0.5, label='Harmonski oscilator')

ax1.set_title('Vrednosti $E_n$, log skala')
ax1.set_xlabel('$n$')
ax1.set_ylabel('$E_n \; (\log) \; [\hbar \omega]$')
ax1.set_yscale('log')
ax1.legend()
ax1.grid()

ax2.plot(n, En, label='Anharmonski oscilator, pad: 0')
ax2.plot(n[:75], En_late, label='Anharmonski oscilator, pad: 25', linestyle='dotted')
ax2.plot(n[:50], En_later, label='Anharmonski oscilator, pad: 50', linestyle='dotted')
ax2.plot(n, n + 0.5, label='Harmonski oscilator')

ax2.set_title('Vrednosti $E_n$, linearna skala')
ax2.set_xlabel('$n$')
ax2.set_ylabel('$E_n \; [\hbar \omega]$')
ax2.set_xlim([0, 200])
ax2.set_ylim([0, 2000])
ax2.legend()
ax2.grid()

plt.tight_layout()
plt.show()
```

Ugotovimo, da lastne vrednosti $E_n$ pri anharmonskem oscilatorju naraščajo z nelinearno odvisnostjo od $N$. Opazimo, da se energijski noviji tudi malo spreminjajo, če vzamemo $n$-ti člen iz Hamiltonjana velikosti $(n+ \texttt{pad} ) \times  (n+ \texttt{pad}) $. Preverimo še kaj se dogaja s posameznim členom $E_n$, ko gre $\texttt{pad} \to \infty$.

```python
E1 = [eig[0][0] for eig in Hs]
E5 = [eig[0][4] for eig in Hs[1:]]
E25 = [eig[0][24] for eig in Hs[5:]]
E50 = [eig[0][49] for eig in Hs[10:]]
E75 = [eig[0][74] for eig in Hs[15:]]

plt.plot(n, E1, label='$E_1$')
plt.plot(n[1:], E5, label='$E_5$')
plt.plot(n[5:], E25, label='$E_{10}$')
plt.plot(n[10:], E50, label='$E_{50}$')
plt.plot(n[15:], E75, label='$E_{75}$')

plt.title('Vrednost $E_n$ v odvisnosti od velikosti $H$')
plt.xlabel('$n + \\tt{pad}$')
plt.ylabel('$E_n \; (\log) \; [\hbar \omega]$')
plt.yscale('log')
plt.legend()
plt.grid()
plt.show()
```

Ugotovimo, da se vrednost $E_n$ numerično sčasoma ustali. Torej lahko izrišemo energijske nivoje, če le vzamemo dovolj veliko vrednost $\tt{pad}$. Iz zgornjega grafa dobimo zelo grobo oceno $\texttt{pad} \approx 10 \cdot n $. Izrišimo še energijske nivoje:

```python
def diagonalize(M):
    _v, _D = np.linalg.eig(M)
    sorting = np.argsort(_v)
    _v = _v[sorting]
    _D = _D[:, sorting]
    return _v, _D

def anharmonic_potential(q, lam=0.5):
    return lam*q**4 + 0.5 * q**2

def harmonic_potential(q):
    return anharmonic_potential(q, lam=0)

def double_well_potential(q):
    return -2*q**2 + 0.1 * q**4

def QHO_eigenFun(q_r, n):
    H = sp.hermite(n)
    return np.power(2**n * sp.factorial(n) * np.sqrt(np.pi), -0.5 ) * np.exp(-0.5 * q_r**2) * H(q_r)

def QHO_lin_comb(q_r, eig_v, i):
    lin_c = np.array([QHO_eigenFun(q_r, n_) for n_ in range(len(eig_v))])
    r = eig_v
    sum = r @ lin_c
    return sum
```

```python
my_lam = 0.5
n = np.arange(1, 500, 5)
H_0s = [H_0(n_) for n_ in n]
Hs = [diagonalize(get_H(n_ + 3, my_lam, 'q')[:n_, :n_]) for n_ in n] # 0 - eigenvalues, 1 - eigenvectors

q_max = 2.3
q_range = np.linspace(-q_max, q_max, 100)
eigen_values = Hs[25][0]
eigen_vectors = Hs[25][1]

plt.plot(q_range, anharmonic_potential(q_range, my_lam), label='Anharmonski potencial', color='black')

for i, eigv in enumerate(eigen_values[:6]):
    plt.axhline(eigv, linestyle='dashed', color='black', linewidth=0.5)
    plt.text(x=-0.4, y=eigv+1.05, s=f'$E_{i} = {eigv:.2f} \, \hbar \omega$', verticalalignment='center', fontsize=8)
    eigen_f = QHO_lin_comb(q_range, eigen_vectors[:, i], i)
    plt.plot(q_range, eigen_f + eigv)
    if i == 0:
        plt.plot(q_range, QHO_eigenFun(q_range, i) + eigv, linestyle='dotted', color='black', linewidth=0.5, label='$| n \\rangle$ za Harmonski oscilator')
    else:
        plt.plot(q_range, QHO_eigenFun(q_range, i) + eigv, linestyle='dotted', color='black', linewidth=0.5)
    plt.fill_between(q_range, eigen_f + eigv, eigv, alpha=0.1)

plt.xlabel('$q$')
plt.ylabel('$E_n \; [\hbar \omega]$')
plt.title("Potencial $V(q) = 0.5\,q^2 + " + str(round(my_lam,2)) + "\,q^4$ in lastne funkcije za $H$")
plt.legend()
plt.show()
```

Iz grafa ugotovimo, da so si lastne funkcije za $H$ (barvne črte), ki so linearna kombinacija lastnih funkcij za harmonski oscilator (črtkana), med sabo precej podobne. Podobnost je še bolj očitna, če namesto verjetnostnih gostot izrišemo kar pričakovane vrednosti. V naslednjem koraku bomo hkrati preučili še limitni primer, ko gre $\lambda \to 0$.
